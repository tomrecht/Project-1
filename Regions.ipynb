```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "combined_data.csv"
all_data = pd.read_csv(file)
```

```python
drop_years = all_data[all_data['Year'] < 2016]
```

```python
drop_years = drop_years.rename(columns = {"Name":"State","Yaddish":"Yiddish"})
```

```python
regions = {
    "West Coast" : ["Alaska","California","Oregon","Washington","Hawaii","Nevada","Arizona"],
    "Gulf Coast" : ["New Mexico","Texas","Arkansas","Louisiana","Mississippi","Alabama"],
    "Rocky Mountains" : ["Montana","Idaho","Wyoming","Utah","Colorado"],
    "Midwest" : ["North Dakota","South Dakota","Nebraska","Kansas","Oklahoma","Minnesota","Iowa","Missouri","Wisconsin","Illinois","Indiana","Michigan","Ohio","Kentucky","Tennessee"],
    "East Coast" : ["Florida","Georgia","South Carolina","North Carolina","Virginia","West Virginia","Maryland","Delaware","New Jersey","Pennsylvania","New York","Vermont","New Hampshire","Massachusetts","Maine","Rhode Island","Connecticut"]
}
```

```python
west_coast = drop_years[drop_years['State'].isin(regions['West Coast'])]
gulf_coast = drop_years[drop_years['State'].isin(regions['Gulf Coast'])]
rocky_mountains = drop_years[drop_years['State'].isin(regions['Rocky Mountains'])]
midwest = drop_years[drop_years['State'].isin(regions['Midwest'])]
east_coast = drop_years[drop_years['State'].isin(regions['East Coast'])]
```

```python
west_coast.to_csv("WestCoast.csv", index = False, header = True)
gulf_coast.to_csv("GulfCoast.csv", index = False, header = True)
rocky_mountains.to_csv("RockyMountains.csv", index = False, header = True)
midwest.to_csv("Midwest.csv", index = False, header = True)
east_coast.to_csv("EastCoast.csv", index = False, header = True)
```

```python
def netChange(df):
    sums2009 = df.loc[df['Year'] == 2009,:].sum()
    sums2015 = df.loc[df['Year'] == 2015,:].sum()
    sums = pd.DataFrame({'2009':sums2009, '2015':sums2015})
    sums = sums.drop(['State','state','Year'])
    sums['Net Change'] = sums['2015'] - sums['2009']
    sums['Percentage'] = sums['Net Change']/sums['2009']*100
    sums_sorted = sums.sort_values(["Net Change"], ascending = False)
    return sums_sorted
```

```python
west_coast_net = netChange(west_coast)
east_coast_net = netChange(east_coast)
gulf_coast_net = netChange(gulf_coast)
rocky_mountains_net = netChange(rocky_mountains)
midwest_net = netChange(midwest)
```

```python
west_coast_net.to_csv("WestCoastNet.csv", index = True, header = True)
gulf_coast_net.to_csv("GulfCoastNet.csv", index = True, header = True)
rocky_mountains_net.to_csv("RockyMountainsNet.csv", index = True, header = True)
midwest_net.to_csv("MidwestNet.csv", index = True, header = True)
east_coast_net.to_csv("EastCoastNet.csv", index = True, header = True)
```

```python
focus_langs = {
    "West Coast" : ["State","Spanish","Chinese","Tagalog","Japanese","Italian","German","Year"],
    "Gulf Coast" : ["State","Spanish","Chinese","Vietnamese","Laotian","Italian","German","Year"],
    "Rocky Mountains" : ["State","Spanish","Chinese","Arabic","Greek","Navajo","German","Year"],
    "Midwest" : ["State","Spanish","Chinese","Arabic","Italian","Polish","German","Year"],
    "East Coast" : ["State","Spanish","Chinese","French Creole","Portuguese","Italian","German","Year"],
}
```

```python
west_coast_focus = west_coast[focus_langs['West Coast']].set_index('Year').groupby('Year').sum()
east_coast_focus = east_coast[focus_langs['East Coast']].set_index('Year').groupby('Year').sum()
gulf_coast_focus = gulf_coast[focus_langs['Gulf Coast']].set_index('Year').groupby('Year').sum()
midwest_focus = midwest[focus_langs['Midwest']].set_index('Year').groupby('Year').sum()
rocky_mountains_focus = rocky_mountains[focus_langs['Rocky Mountains']].set_index('Year').groupby('Year').sum()
```

```python
def percentage(df):
    for lang in df.columns:
        baseline = df[lang][2009]
        df[lang] = df[lang] / baseline
    return df
```

```python
plt.figure(figsize=(20, 12))
plt.plot(percentage(midwest_focus))
plt.title("Change in selected languages, Midwest region (2009 = 100%)")
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
plt.legend(midwest_focus.columns)
plt.rc('figure',titlesize=20)
plt.rc('axes',titlesize=30)
plt.savefig("Midwest.png")
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(percentage(east_coast_focus))
plt.title("Change in selected languages, East Coast region (2009 = 100%)")
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
plt.legend(east_coast_focus.columns)
plt.rc('figure',titlesize=20)
plt.rc('axes',titlesize=30)
plt.savefig("EastCoast.png")
plt.show()
plt.clf()
```
