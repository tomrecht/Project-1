```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census_2009 = pd.read_csv('2009.csv')
census_2009 = census_2009.iloc[:,:-1]
census_2009.head()
```

```python
census_2009 = pd.read_csv('2009.csv')
census_2009 = census_2009.iloc[:,:-1]
tot_2009 = pd.DataFrame(census_2009.sum(numeric_only=True)).T
tot_2009.index = ["2009"]
tot_2009
```

```python
census_2018 = pd.read_csv('2018.csv')
census_2018 = census_2018.iloc[:,:-1]
tot_2018 = pd.DataFrame(census_2018.sum(numeric_only=True)).T
tot_2018.index = ["2018"]
tot_2018
```

```python
combined = tot_2009.append(tot_2018)
combined
```

```python
combined = tot_2009.append(tot_2018).T
#combined.set_index('Language')
combined.index.name = 'Language'
combined.head()
```

```python
combined['Change'] = combined['2018'] - combined['2009']
combined['Percentage'] = combined['Change']/combined['2009']*100
#combined = combined.sort('Percentage', ascending=False).head(5)
combined = combined.sort_values(["Percentage"], ascending = False).style.format({"Percentage":"{:.2f}%"})
combined
```
