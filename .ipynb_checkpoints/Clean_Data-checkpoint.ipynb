```python
# function to access census data from 2009 to 2015, and save data into csv files individually
def census(c,year):
    census_data = c.acs5.get(("NAME", "B16001_002E", "B16001_003E", "B16001_006E",
                          "B16001_021E","B16001_009E","B16001_012E","B16001_015E","B16001_018E",
                          "B16001_030E","B16001_033E","B16001_036E","B16001_039E","B16001_045E",
                          "B16001_048E","B16001_051E","B16001_054E","B16001_057E","B16001_066E",
                          "B16001_069E","B16001_072E","B16001_075E","B16001_078E","B16001_081E",
                          "B16001_084E","B16001_087E","B16001_093E","B16001_099E","B16001_105E",
                          "B16001_105E","B16001_108E","B16001_111E"), {'for': 'state:*'})
    census_pd = pd.DataFrame(census_data)
    census_pd = census_pd.rename(columns={"NAME": "Name",
                                          "B16001_002E":"English only", 
                                          "B16001_003E":"Spanish", 
                                          "B16001_006E":"French",
                                          "B16001_021E":"Yaddish",
                                          "B16001_009E": "French Creole",
                                          "B16001_012E": "Italian",
                                          "B16001_015E": "Portuguese",
                                          "B16001_018E": "German",
                                          "B16001_030E": "Greek",
                                          "B16001_033E": "Russian",
                                          "B16001_036E": "Polish",
                                          "B16001_039E": "Serbo-Croatian",
                                          "B16001_045E": "Armenian",
                                          "B16001_048E": "Persian",
                                          "B16001_051E": "Gujarati",
                                          "B16001_054E": "Hindi",
                                          "B16001_057E": "Urdu",
                                          "B16001_066E": "Chinese",
                                          "B16001_069E": "Japanese",
                                          "B16001_072E": "Korean",
                                          "B16001_075E": "Mon-Khmer",
                                          "B16001_078E": "Hmong",
                                          "B16001_081E": "Thai",
                                          "B16001_084E": "Laotian",
                                          "B16001_087E": "Vietnamese",
                                          "B16001_093E": "Tagalog",
                                          "B16001_099E": "Navajo",
                                          "B16001_105E": "Hungarian",
                                          "B16001_108E": "Arabic",
                                          "B16001_111E": "Hebrew"})
    census_pd.to_csv(f"output_data/{year}_census.csv",index = None, header=True)
```

```python
# Print out each year's census data individually
years = [2009,2010,2011,2012,2013,2014,2015]
for year in years:
    c = Census(census_api_key, year)
    census(c,year)
```

```python
# Mix all files into one file
census_files = sorted(glob("output_data/*_census.csv"))
census_files
```

```python
# Change the file name that appear on the last column to only the year
census_df = pd.concat((pd.read_csv(file).assign(Year=file) 
                       for file in census_files),ignore_index=True)

census_df = census_df.replace("output_data/2009_census.csv","2009")
census_df = census_df.replace("output_data/2010_census.csv","2010")
census_df = census_df.replace("output_data/2011_census.csv","2011")
census_df = census_df.replace("output_data/2012_census.csv","2012")
census_df = census_df.replace("output_data/2013_census.csv","2013")
census_df = census_df.replace("output_data/2014_census.csv","2014")
census_df = census_df.replace("output_data/2015_census.csv","2015")

census_df = census_df.sort_values(["Name","Year"])
census_df = census_df.reset_index(drop=True)
census_df.head(10)
```

```python
# Save the csv file
census_df.to_csv("combined_data.csv",index = None, header=True)
```
