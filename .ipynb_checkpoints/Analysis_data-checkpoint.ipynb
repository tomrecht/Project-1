```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census_data = pd.read_csv("combined_data.csv")
census_data
```

## Find out the percentage change on every states

```python
state_change = (census_data.groupby(['Name']).pct_change()*100).round(2)
name = census_data['Name'].tolist()
state_change["Name"] = name
state_change = state_change.reindex(columns = ['Name','English only', 'Spanish', 'French Creole', 'Yaddish', 
        'Italian','Portuguese', 'German', 'Greek', 'Russian', 'Polish', 'Serbo-Croatian',
       'Armenian', 'Persian', 'Gujarati', 'Hindi', 'Urdu', 'Chinese',
       'Japanese', 'Korean', 'Mon-Khmer', 'Hmong', 'Thai', 'Laotian',
       'Vietnamese', 'Tagalog', 'Navajo', 'Hungarian', 'Arabic', 'Hebrew',
       'state', 'Year'])
del state_change["state"]
del state_change["Year"]
state_change = state_change.fillna(0)
state_change
```

```python
state_change.to_csv("state_change.csv",index = None, header=True)
```

## Percentage change group by year

```python
yearly_sum = census_data.groupby(['Year']).sum()
del yearly_sum['state']
yearly_sum
```

```python
yearly_sum.to_csv("yearly_sum.csv",index = None, header=True)
```

```python
yearly_change = (yearly_sum.pct_change()*100).round(2)
yearly_change = yearly_change.drop(yearly_change.index[0])
yearly_change
```

```python
yearly_change.to_csv("yearly_change.csv",index = None, header=True)
```
