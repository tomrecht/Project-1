```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census_data = pd.read_csv('SumsByYear.csv')
census_data = census_data.iloc[:,:8]
census_data.rename(columns = {'Unnamed: 0':'Language'}, inplace = True)
census_data.head()
```

```python
navajo = census_data[census_data["Language"] == "Navajo"]
#cap_mouse = cap[cap["Mouse ID"] == "b128"]
navajo = navajo.loc[:,:]
navajo.set_index(['Language'], inplace=True)
navajo_rev = navajo.T
navajo_rev
```

```python
navajo_rev.plot(kind="bar",  rot=45, figsize=(7,3))
plt.title("The Navajo - Speaking Population from 2009 to 2015")
plt.ylim(150000, 175000)
plt.show()
plt.tight_layout()
plt.savefig("Fig1.png")
```

```python
census_2015 = pd.read_csv('2015.csv')
census_2015 = census_2015.iloc[:,:-1]
census_2015.head()
```

```python
census_2015.rename(columns = {'Name':'State'}, inplace = True)
census_navajo = census_2015.loc[:, ["State","Navajo"]]
census_navajo = census_navajo.sort_values(by='Navajo', ascending=False)
#navajo_summary = census_navajo.iloc[0:5]
census_navajo.set_index(['State'], inplace=True)
census_navajo = census_navajo.iloc[0:5]
census_navajo
```

```python
census_navajo.plot(kind="bar",  rot=45, figsize=(7,3))
plt.title("The five states use the Navajo language the most in 2015")
plt.show()
plt.tight_layout()
plt.savefig("Fig2.png")
```
